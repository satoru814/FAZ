```python
import shapely
import matplotlib.pyplot as plt
import json
import numpy as np
import os
import cv2
from shapely.geometry import Polygon
from shapely.geometry import Point


path = "./FAZ_sgementation/superficial_all/1.json"
with open(path, "r") as f:
    jso = json.load(f)
a = jso["shapes"][0]["points"]
a = Polygon(a)
img = cv2.imread(path.replace(".json", ".jpg"))
mask = np.zeros((img.shape[0], img.shape[1]))
for h in range(img.shape[0]):
    for w in range(img.shape[1]):
        b = a.contains(Point(w,h))
        if b :
            mask[h,w] = 1
```

```python
path = "./FAZ_sgementation/superficial_all_png/3_mask.png"
img = cv2.imread(path, )
```

```python
fig,ax = plt.subplots(1,2)

ax[0].imshow(mask)
ax[0].set_title("mask image")
ax[1].imshow(img)
ax[1].set_title("image")
plt.savefig("./assets/mask-and-img.png")
```
